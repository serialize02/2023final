### BeautifulSoup

html에서 텍스트만 추려오는 법

```python
from bs4 import BeautifulSoup
```

```python
with open("./myFirstWeb.html", "r") as f:
  content = f.read()
  soup = BeautifulSoup(content, "lxml")
```

<h4>Head</h4>

```python
print(soup.head)
```

```python
print(soup.head.title)
```

```python
print(soup.head.title.get_text())
```

꺽쇠 빼고 나머지

<h4>Text</h4>

```python
paragraphs = soup.find_all("p")
```

```python
for p in paragraphs:
  print(p.get_text())
```

### Practice

```python
response = requests.get("https://en.wikipedia.org/wiki/Cat")

soup = BeautifulSoup(response.content, "lxml")
```

```python
paragraphs = soup.find_all("p")

for p in paragraphs:
  print(p.get_text())
```

```python
links = soup.find_all("a")
```

```python
for link in links:
  url = link.get("href")

  if url != None and url.startswith("https"):
    print(link.get("href"))
```
